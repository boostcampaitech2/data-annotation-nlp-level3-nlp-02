# text-link dictionary 생성

```python
import os
import re
import kss
import json
import glob
from tqdm import tqdm
```

```python
if os.path.exists("./all_text.json"):
    with open("./all_text.json","r") as f:
        all_text = json.load(f)
else:
    post_url = "https://ko.wikipedia.org/wiki/"
    all_text = dict()
    for file in tqdm(glob.glob("./text/*.txt")):
        filename = re.sub("./text/|.txt","",file)
        with open(file, "r",) as f:
            strings = f.readlines()
        tmp_sents = []
        for s in strings:
            s = re.sub("\n","",s)
            if s != "":
                tmp_sents += kss.split_sentences(s)
                url_link = filename.replace(" ","_")
                url_link = post_url + url_link
        all_text[filename] = {"text" : tmp_sents, "link" : url_link}

    with open("./all_text.json", "w") as f:
        json.dump(all_text,f,ensure_ascii=False)
```

```python
from collections import defaultdict
tmp_dict = defaultdict(int)
for key in all_text.keys():
    d_ = all_text[key]['text']
    i = 1
    for s in d_:
        print(f"[{i}, {key}] {s}")
        i+=1
            # tmp_dict[key] +=1
```
